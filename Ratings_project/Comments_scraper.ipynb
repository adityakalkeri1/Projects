```python
from selenium import webdriver
from webdriver_manager.firefox import GeckoDriverManager
import time
import csv
from selenium.webdriver.common.keys import Keys
```

```python
def amazon__comments(product_name):
    ratings_dict = {5:1,4:2,3:3,2:4,1:5}
    csv_file = open('amazon_{}.csv'.format(product_name), 'w')
    writer = csv.writer(csv_file)
    writer.writerow(['Comment', 'Rating'])
    driver = webdriver.Firefox(executable_path = GeckoDriverManager().install())
    driver.get('https://www.amazon.in')
    time.sleep(5)
    driver.execute_script('window.open()')
    driver.find_element_by_xpath('//*[@id="twotabsearchtextbox"]').send_keys(product_name)
    driver.find_element_by_xpath('//*[@id="nav-search-submit-button"]').click()
    #First Loop: to loop over the individual products
    for page in range(20):
        print('****Page {}****'.format(page))
        for prod in range(20):
            try:
                try:
                    url = driver.find_element_by_xpath('/html/body/div[1]/div[2]/div[1]/div/div[1]/div/span[3]/div[2]/div[{}]/div/span/div/div/div[2]/div[2]/div/div[1]/h2/a'.format(prod)).get_attribute('href')
                    print(url)
                except:
                    try:
                        url = driver.find_element_by_xpath('/html/body/div[1]/div[2]/div[1]/div/div[1]/div/span[3]/div[2]/div[{}]/div/div/div/div[2]/div[2]/div/div[1]/h2/a'.format(prod)).get_attribute('href')
                        print(url)
                    except:
                        continue
                print('Product No: {}'.format(prod))
                driver.switch_to.window(driver.window_handles[1])
                driver.get(url)
                print('Page opened 1')
                time.sleep(10)
            #Opening comments page
                try:
                    com_url = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[5]/div[37]/div/div/div[2]/div/div[2]/span[3]/div/div/div[5]/div[2]/a').get_attribute('href')
                except:
                    com_url = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[5]/div[26]/div/div[2]/div/div[2]/span[3]/div/div/div[5]/div[2]/a').get_attribute('href')
                driver.get(com_url)
                print('comments page opened')
                time.sleep(10)
            #selecting no of stars
                for stars in range(1,6):
                    driver.find_element_by_xpath('//*[@id="a-autoid-6-announce"]').click()
                    time.sleep(1)
                    driver.find_element_by_xpath('//*[@id="star-count-dropdown_{}"]'.format(stars)).click()
                    time.sleep(5)
                    print('filter applied')
                    comments_count = 0
                    condition = True
                    while condition:
                        for i in range(2,11):
                            try:
                                text= driver.find_element_by_xpath('/html/body/div[1]/div[3]/div/div[1]/div/div[1]/div[5]/div[3]/div/div[{}]/div/div/div[4]/span'.format(i)).text
                                print(text)
                                rating = ratings_dict[stars]
                                print(rating)
                                writer.writerow([text, rating])
                            
                                comments_count = comments_count+1
                                print(comments_count)
                            except:
                                continue
                        if comments_count>75:
                            condition = False
                        else:
                            try:
                                driver.find_element_by_xpath('/html/body/div[1]/div[3]/div/div[1]/div/div[1]/div[5]/div[3]/div/div[12]/span/div/ul/li[2]/a').click()
                                time.sleep(5)
                            except:
                                condition = False
                driver.switch_to.window(driver.window_handles[0])
            except:
                continue
        
        driver.switch_to.window(driver.window_handles[0])
        time.sleep(5)
        next_page = driver.find_element_by_css_selector('li.a-last > a:nth-child(1)').get_attribute('href')
        driver.get(next_page)
        time.sleep(5)
```

```python
def flipkart_comments(product_name):
    csv_file = open('flipkart_{}.csv'.format(product_name), 'w')
    writer = csv.writer(csv_file)
    writer.writerow(['Comment', 'Rating'])
    driver = webdriver.Firefox(executable_path=GeckoDriverManager().install())
    driver.get('https://www.flipkart.com')
    driver.execute_script('window.open()')
    time.sleep(5)
    driver.find_element_by_xpath('/html/body/div[2]/div/div/button').click()
    driver.find_element_by_xpath('/html/body/div/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input').send_keys(product_name)
    driver.find_element_by_xpath('/html/body/div/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input').send_keys(Keys.ENTER)
    time.sleep(5)
    page_condition = True
    while page_condition:
        for product in range(1,26):
            try:
                url = driver.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div[{}]/div/div/div/a'.format(product)).get_attribute('href')
                print(url)
                driver.switch_to.window(driver.window_handles[1])
                driver.get(url)
                print('Product_page_opened')
                time.sleep(5)
                driver.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div[9]/div[7]/div/a/div').click()
                print('Comments_page_opened')
                time.sleep(5)
                #comment_condition = True
                comments = 0
                while comments <250:
                    for i in range(1,15):
                        try:
                            text = driver.find_element_by_xpath('/html/body/div[1]/div/div[3]/div/div[1]/div[2]/div[{}]/div/div/div/div[2]/div/div/div'.format(i)).text
                            print(text)
                            rating = driver.find_element_by_xpath('/html/body/div[1]/div/div[3]/div/div[1]/div[2]/div[{}]/div/div/div/div[1]/div'.format(i)).text
                            print(rating)
                            writer.writerow([text, rating])
                            print('#####################')
                            comments = comments+1
                            print(comments)
                        except:
                            continue
                    driver.find_element_by_xpath('/html/body/div[1]/div/div[3]/div/div/div[2]/div[13]/div/div/nav/a[11]').click()
                    print('######################Comments Page next###################')
                    time.sleep(5)
            except:
                pass
            driver.switch_to.window(driver.window_handles[0])
            print('Product {} Done'.format(product))
            
        print('##############################################')
        print('Page Done, Opening next page')
        print()
        driver.switch_to.window(driver.window_handles[0])
        driver.find_element_by_xpath('/html/body/div/div/div[3]/div[1]/div[2]/div[26]/div/div/nav/a[11]').click()
        print('#############NEXT PAGE OPENED#################')
```
